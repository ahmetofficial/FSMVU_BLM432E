```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()

languages=['Python','SQL','Java','C++','JavaScript']
pos = np.arange(len(languages))
popularity=[56,39,34,34,29]

bars= plt.bar(pos,popularity,align='center',width=0.3,color='gray')

#change color
bars[0].set_color("yellow")

#write title
plt.title("Programming Language Usage")

#write the languages on the x-axis
plt.xticks(pos,languages)
plt.yticks([])

#remove ticks on x-axis
ax = plt.axes()
ax.xaxis.set_ticks_position('none') 

#remove ticks on y-axis
ay = plt.axes()
ay.yaxis.set_ticks_position('none')

#remove frame
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)


#add label y-values
for rect in bars:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2., height-5,'%d' % int(height),ha='center', 
            va='bottom')



plt.show()
```
